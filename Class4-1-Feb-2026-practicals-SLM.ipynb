# Use SLM Distilled GPT2 (Distilgpt2)  and (tinygpt2) 

```python
import tensorflow as tf
import numpy as np
from transformers import pipeline, set_seed
```

```python
# 1. Initialize the pipeline with 'distilgpt2'
# This is the standard SLM version of GPT-2
#Set a seed so that same results are re-produced (to set the random state)
set_seed(42)
slm_generator = pipeline(
    task = "text-generation",     #task of SLM is text-generation
    model="distilgpt2",           #model identified
    device=-1)                    #device = -1 is then CPU used, device = 0 then GPU used
```

```python
#set the prompt you will pass
prompt = "Newton's first law says that"
```

```python
# 2. Generate the response
# Using max_new_tokens ensures we get fresh text regardless of prompt length
slm_generator.tokenizer.pad_token = slm_generator.tokenizer.eos_token
output = slm_generator(
    prompt, 
    max_length=100,       # Maximum length of Prompt + Response 
    max_new_tokens=50,    # how many new tokens can be added
    temperature=0.1,      # higher value of 0.7 means more creative
    top_p = 0.9,           # the set of the tokens with cumulative probability of top p 
    num_return_sequences=3, # how many responses are required
#   truncate=True,          #truncate
    pad_token_id=slm_generator.tokenizer.eos_token_id, # Sets the padding character
    padding=True,                                       # Enables padding logic
    do_sample=True        # enable more creative sampling
)
```

```python
# 3. Extract and print line by line
for i, out in enumerate(output, start=1):
    print(f"\n ------ Output {i} ------")
    print(out['generated_text'])
```
